#### [실습문제3] 데이터 확인 및 처리

* 데이터를 보았을때, 어느 데이터를 가져와야 할지 생각 해 봅시다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
a=seoul_business.loc[seoul_business["동"]=='소계']
a=a.drop("동",axis=1)
a.set_index('자치구',inplace=True)
for i in a.columns:
    y=a[i].values
    u=[]
    for j in y:
        j=int(j.replace(",",""))
        u.append(j)
    a[i]=u
# a.astype('int')   
a.reset_index(inplace=True)
a.info()
a
```
